### Step 1. 필요한 라이브러리 들을 import 하세요

### Step 3. iris 라는 변수에 DataFrame 을 할당하세요

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url)
```

### Step 4. 다음과 같은 순서로 columns을 만드세요

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class
iris.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
```

```python
iris.head()
```

### Step 5. 데이터 프레임의 각컬럼 마다 누락된 값이 존재 합니까?

```python
iris.isnull().sum()
iris.info()
```

### Step 6. 'petal_length' 열의 10~29행 값을 `NaN`으로 설정하세요

```python
import numpy as np
```

```python
type(np.nan)
```

```python
iris.iloc[10:30,2] = np.nan
```

```python
iris.head(20)
```

### Step 7. `NaN` 값을 1.0 으로 대체 하세요

```python
iris.petal_length.fillna(1, inplace=True)
```

```python
iris.head(20)
```

### Step 8. class column을 삭제해 보세요

```python
del iris["class"]
iris.head()
```

### Step 9. 처음 3개 행을 NaN으로 설정하세요

```python
# iris.iloc[:3] = np.nan
iris.iloc[:3, :] = np.nan # column을 명확하게 지정하는게 좋습니다. 
```

```python
iris.head()
```

### Step 10. NaN값이 있는 행을 삭제 하세요

```python
iris.dropna(inplace=True)
iris.head()
```

### Step 11. index 를 0 부터 시작하도록 reset하세요

```python
iris.reset_index()
# iris.reset_index(drop=True, inplace=True) # drop=True 기존의 index를 제거
```
