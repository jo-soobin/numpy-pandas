### Step 3. users 변수에 DataFrame을 할당 하세요.
#### index 는 user_id입니다.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep="|", index_col="user_id")
```

```python
users.head()
```

### Step 4.  occupation별 평균 나이는 얼마입니까?

```python
users.groupby("occupation").age.mean()
```

### Step 5. occupation별 남성의 비율을 확인하고 내림차순으로 정렬하세요

```python
set(users.gender)
```

```python
def gender2num(x):
  if x == "M":
    return 1
  if x =="F":
    return 0

users["gender2num"] = users["gender"].apply(gender2num)
```

```python
mm_df = users.groupby("occupation").gender2num.sum() / users.occupation.value_counts() *100
mm_df.sort_values(ascending=False)
```

```python
# 다른풀이
(users[users.gender == 'M'].groupby('occupation').count()['gender'] / users.groupby('occupation').count()['gender']).sort_values(ascending = False)
```

### Step 6. 각 occupation별 최대 최소 age를 확인하세요

```python
users.groupby("occupation").age.agg(["max", "min"])
```

### Step 7. occupation에 따른 gender별 평균 age를 계산 하세요

```python
users.groupby(["occupation", "gender"]).age.mean()
```

### Step 8. 각 occupation별 남성과 여성의 비율을 나타내시오
#### 어려움

```python
gender_oc = users.groupby(["occupation","gender"]).agg({"gender":"count"})
```

```python
oc_count = users.groupby(["occupation"]).count()
```

```python
oc_count
```

```python
gender_oc.div(oc_count, level = "occupation") # 
```
