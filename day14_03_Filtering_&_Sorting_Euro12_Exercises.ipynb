```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url, sep=",")
```

### Step 7. discipline 변수에 각 Team 당 Yellow Cards, Red Cards 을 dataframe으로 할당하여라

```python
discipline = euro12[["Team","Yellow Cards", "Red Cards"]]
discipline.head()
```

### Step 8. Red Cards로 정렬한 다음 Yellow Cards로 정렬하시오

```python
discipline.sort_values(["Red Cards", "Yellow Cards"], ascending=False)
```

### Step 9. 팀당 평균 Yellow Cards 의 비율은 얼마인가?

```python
# 전체팀이 평균 몇개의 yellow cards를 받았는지 계산
round(discipline["Yellow Cards"].mean())
```

### Step 10. 6골이상 득점한 팀을 필터링 하시오

### Step 11."G" 로 시작하는 팀을 선택하시오

```python
euro12.Team.str.startswith("G")
```

```python
euro12[euro12['Team'].str[0] == 'G']
```

```python
def start_str(str_1):
  if str_1[0] == "G":
    return True
  else:
    return False
euro12[euro12.Team.apply(start_str)]
```
