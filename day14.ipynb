```python
import pandas as pd
```

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|') #, index_col=['user_id', "gender"]
# index_col : index로 지정할 column을 지정 index_col=['user_id', "gender"] multi_index 지정 가능
```

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', header = 1)
#header : column의 이름으로 지정할 행 지정
# header = None : column name없음
# header = 0 : 첫번째 행을 column name으로 지정
# header = 0 : 두번째 행을 column name으로 지정
```

```python
col_names = ["col1", "col2", "col3", "col4", "col5" ]
pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', names = col_names)
```
